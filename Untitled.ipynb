```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

```python
train.head()
```

```python
train.columns = [x.lower() for x in train.columns]
```

```python
f, axs = plt.subplots(2,2,figsize=(20,15))
f.suptitle('Analysing',y=0.93,fontsize=16)

## histogram of saleprice

sns.distplot(train.saleprice, kde=False, ax=axs[0,0])
axs[0,0].set_title('Histogram of SalePrice')


## correlation map

sns.heatmap(train[['saleprice','lotfrontage','lotarea','totalbsmtsf','fireplaces','garagearea',
                   'poolarea','fullbath','1stflrsf','2ndflrsf','grlivarea','garagecars']].corr(),cmap='Reds',vmin=-1,vmax=1,annot=True,ax=axs[0,1])
axs[0,1].set_title('Correlation Map')

## box plot of mssubclass

sns.boxplot(x='mssubclass',y='saleprice',data=train,ax=axs[1,0])
axs[1,0].set_title('Boxplot of MSSubclass')

## boxplot of overallqual

sns.boxplot(x='overallqual',y='saleprice',data=train,ax=axs[1,1])
axs[1,1].set_title('Boxplot of OverallQuality');
```

```python
corrmat =train[['saleprice','lotfrontage','lotarea','totalbsmtsf','fireplaces','garagearea','poolarea','fullbath','1stflrsf','2ndflrsf','grlivarea','garagecars']].corr() 

features = list(corrmat.index[abs(corrmat['saleprice'] > 0.45)])

features.extend(['mssubclass','overallqual','salecondition'])
```

```python
num_features = ['totalbsmtsf','fireplaces','garagearea','fullbath','1stflrsf','grlivarea','garagecars','saleprice']
cat_features = ['mssubclass','overallqual','salecondition']

df = train[features].copy()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.isna(),cbar=False)
plt.show()
```

```python
sns.pairplot(df[num_features]);
```

```python
skewness = df[num_features].apply(lambda x: skew(x))
skewness.sort_values(ascending=False)
```
